## The Code

### Imports

```python
import pandas as pd
```

### Load the dataset into a DataFrame

```python
# Load the dataset into a DataFrame
def load_dataset():
    # TODO: Load the dataset from a CSV file or define it here as a dictionary and create a DataFrame.
    
    # Define the dataset as a Pandas DataFrame
    data = {
        'Plant ID': [1, 2, 3, 4, 5, 6, 7, 8],
        'Leaf Shape': ['Lobed', 'Lobed', 'Oval', 'Oval', 'Palmate', 'Palmate', 'Oval', 'Oval'],
        'Flower Color': ['Pink', 'Pink', 'Yellow', 'Yellow', 'White', 'White', 'Red', 'Red'],
        'Species': ['Rose', 'Cherry Blossom', 'Daffodil', 'Sunflower', 'Magnolia', 'Dogwood', 'Tulip', 'Poppy'],
        'Certainity Factor(Shape,color)': [(0.7, 0.8), (0.7, 0.7), (0.8, 0.9), (0.8, 0.85), (0.75, 0.7), (0.75, 0.75), (0.8, 0.8), (0.8, 0.75)]
    }
    df = pd.DataFrame(data)
    return df
```

### Identify the plant based on user input

```python
# Define a function to identify the plant based on user input
def identify_plant(df, user_shape, user_color):
    max_certainty = 0
    identified_plant = None

    for index, row in df.iterrows():
        if row['Leaf Shape'] == user_shape or row['Flower Color'] == user_color:
            # Calculate the overall certainty factor
            overall_certainty = min(row['Certainity Factor(Shape,color)'])
            
            # Update identified plant if the certainty is higher
            if overall_certainty > max_certainty:
                max_certainty = overall_certainty
                identified_plant = row['Species']
                
    return identified_plant, max_certainty
```

### Get user input and identify plants


```python
# Function to get user input and identify plants
def plant_identification(df):
    user_shape = input("Enter the leaf shape: ").strip().title()
    user_color = input("Enter the flower color: ").strip().title()

    identified_plant, max_certainty = identify_plant(df, user_shape, user_color)

    if identified_plant:
        print(f"The most likely plant species is {identified_plant} with a certainty factor of {max_certainty}")
    else:
        print("No matching plant species found.")
```

### Main function to solve the problem

```python
# Main program
if __name__ == "__main__":
    print("Plant Identification System")
    df = load_dataset()
    plant_identification(df)
```
